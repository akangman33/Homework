```python
import requests
```

```python
res = requests.get("https://movies.yahoo.com.tw/movieinfo_main/%E6%80%AA%E7%8D%B8%E8%88%87%E8%91%9B%E6%9E%97%E6%88%B4%E8%8F%AF%E5%BE%B7%E7%9A%84%E7%BD%AA%E8%A1%8C-fantastic-beasts-the-crimes-of-grindelwald-7963?guccounter=1")
```

```python
res.text
```

```python
from bs4 import BeautifulSoup as bs
import re
```

```python
soup = bs(res.text, "lxml")
```

```python
movie = []
movie_cnname = soup.select_one(".movie_intro_info_r h1").text
movie_egname = soup.select_one(".movie_intro_info_r h3").text
movie_type = []
for i in soup.select(".level_name a"):
    movie_type.append(i.text.strip())
movie_date = soup.select(".movie_intro_info_r span")[0].text
movie_length = soup.select(".movie_intro_info_r span")[1].text
movie_company = soup.select(".movie_intro_info_r span")[2].text
movie_imdb = soup.select(".movie_intro_info_r span")[3].text
movie_director = soup.select(".movie_intro_list")[0].text.strip()
movie_actor = soup.select(".movie_intro_list")[1].text
movie_plot = soup.select_one(".gray_infobox_inner span").text
movie.append(movie_cnname)
movie.append(movie_egname)
movie.append(movie_type)
movie.append(movie_date)
movie.append(movie_length)
movie.append(movie_company)
movie.append(movie_imdb)
movie.append(movie_director)
movie.append(movie_actor)
movie.append(movie_plot)
```
